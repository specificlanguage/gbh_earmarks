```python
import pandas as pd
```

```python
data_per_year = []
for yr in range(2010, 2021):
    data_per_year.append(pd.read_csv("data/ma_census/acs_rent_nb/" + str(yr) + ".csv", header=1))
print(data_per_year)
```

```python
burdened_percentages = []
x = -1
burdened = pd.DataFrame()

df = data_per_year[0]
tdf = df.iloc[0, 13:21:2]
num_burdened = int(tdf.sum())
total = int(df.iloc[0, 1])

burdened = burdened.append([[2010, total, num_burdened, num_burdened / total]])

print(burdened)

for i in range(1, len(data_per_year)):
    df = data_per_year[i]
    tdf = df.iloc[0, 12:20:2]
    num_burdened = tdf.sum()
    total = df.iloc[0, 0]
    burdened = burdened.append([[2010 + i, total,  num_burdened, num_burdened / total]])

burdened = burdened.rename(columns={0: "year", 1: "total", 2: "number_burdened", 3: "pct_burdened"})
burdened.to_csv(columns=burdened.columns, path_or_buf="out/nb_burdened_per_year.csv")
```
